```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np
from math import *
```

```python
L=0.1
m=0.08
```

```python
M = (m/420)*np.array([[156, -22*L],[-22*L, 4*L**2]])
M
```

```python
A = np.array([[1, 1],[13.775, 76.225]])
A
```

```python
M_star = A.T@M@A
M_star
```

```python
f = np.array([1,0])
f_star = A.T@f
f_star
```

```python
omega1 = 5082.715
omega2 = 50078.407
```

```python
t = np.linspace(0, 0.005, 500)
damping_ratio = 0.05
```

```python
lambda1 = f_star[0]*(1/(M_star[0,0]*omega1*sqrt(1-damping_ratio**2)))*np.exp(-damping_ratio*omega1*t)*np.sin(omega1*sqrt(1-damping_ratio**2)*t)
```

```python
lambda2 = f_star[1]*(1/(M_star[1,1]*omega2*sqrt(1-damping_ratio**2)))*np.exp(-damping_ratio*omega2*t)*np.sin(omega2*sqrt(1-damping_ratio**2)*t)
```

```python
plt.plot(t, lambda1, label='1 mode')
plt.plot(t, lambda1+lambda2, label='2 modes')
plt.xlabel('Time (sec)')
plt.ylabel('Displacement (m)')
plt.title('Impulse response of a cantilever beam using modal superposition')
plt.legend(loc='best')
```
